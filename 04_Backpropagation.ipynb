## **Local gradients**

dz/dx = dx.y/dx = y

dz/dy = dx.y/dy = x


in training we want to calculate the gradient of the loss with the respect to the x:

dLoss/dx = dLoss/dz. dz/dx


## **Concepts:**

first step: Forward pass : compute loss

second step: Compute local gradients

third step: Backward pass: compute dLoss/dWeights using the chain rule

```python
import torch
```

```python
x = torch.tensor(1.0)
y = torch.tensor(2.0)

w = torch.tensor(1.0, requires_grad=True)

#forward pass and compute the loss
y_hat = w * x
loss = (y_hat - y)**2

print(loss)
```

```python
#backward pass
loss.backward()
print(w.grad)
```
